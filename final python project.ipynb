```python
def condition(salary):
    if salary<=20000:
        return'Minimum'
    elif 20000<=salary<=50000:
        return'Average'
    elif 50000<=salary<=100000:
        return'Maximum'
    else:
        return 'NA'
```

```python
bank['salary_catg']=bank['salary'].apply(condition)
```

```python
bank.head()
```

```python
bank['salary_catg'].value_counts()
```

```python
bank.groupby(['salary_catg','response']).campaign.mean().plot(kind='bar')
```
